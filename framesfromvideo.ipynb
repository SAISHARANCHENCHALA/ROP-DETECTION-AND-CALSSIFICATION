```python
import cv2

# path of video
vidcap = cv2.VideoCapture("D:\\BTP\\AROP2\\VID7\\5215db48-76b6-459c-acdb-0adf383c8c34.2-002.avi")

if not vidcap.isOpened():
    print("Error: Could not open video capture object")
else:
    count = 0
    while True:
        success, image = vidcap.read()
        if success:
            cv2.imwrite("D:\\BTP\\AROP2\\VID7\\%d.png" % count, image)
            count += 1
            print(f"Extracted frame {count}")
        else:
            print("Finished extracting frames")
            break
```
